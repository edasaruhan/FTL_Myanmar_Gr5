```python
import ast, re, numpy as np, pandas as pd

def extract_safety_value(s):
    if isinstance(s, dict):
        return s.get("Hallucination", np.nan)
    if pd.isna(s):
        return np.nan
    try:
        d = ast.literal_eval(s)
        if isinstance(d, dict):
            return d.get("Hallucination", np.nan)
    except Exception:
        pass
    m = re.search(r"""["']?Hallucination["']?\s*:\s*([0-9]+(?:\.[0-9]+)?)""", str(s))
    return float(m.group(1)) if m else np.nan

df = pd.read_csv("medical_test_results.csv")
col_name = "evaluation" if "evaluation" in df.columns else "eval"
df["Hallucination"] = df[col_name].apply(extract_safety_value)

print(df["Hallucination"].describe())
df[["input","model_answer","Hallucination"]].head(10)
```
