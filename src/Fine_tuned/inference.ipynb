```python
from simpletransformers.t5 import T5Model

model = T5Model("mt5", "YeBhoneLin10/FTL-Capstone-v2",device=0)
```

```python
import pandas as pd 

test_dataset11 = pd.read_csv("mental_health.csv")
test_dataset11 = test_dataset11[0:5000]
```

```python
import pandas as pd
from transformers import AutoTokenizer
import torch
from tqdm import tqdm

tokenizer = model.tokenizer
hf_model = model.model


predictions = []

for row in tqdm(test_dataset11.itertuples(), total=len(test_dataset11)):
    asr_input = getattr(row, 'input')
    input_text = f"{asr_input}"

    inputs = tokenizer(input_text, return_tensors="pt").to(hf_model.device)

    outputs = hf_model.generate(
        inputs.input_ids,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.9,
        num_return_sequences=1,
        max_length=256,
    )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    predictions.append(decoded)

test_dataset11["model_answer"] = predictions

test_dataset11.to_csv("evaluation.csv", index=False)
```
